```python
import aisuite as ai
```

AISuite works as a middle layer between your code and many LLM providers, exposing one OpenAI‑style API and internally routing each call to the right provider, handling auth, parameters, tools, and optional agent loops for you. Instead of learning every provider SDK, you talk only to aisuite.Client, and it decides which provider module to use based on the model="provider:model-name" string you pass.

Setup AWS access

The Client class in aisuite serves as the central interface for interacting with multiple LLM providers through a unified OpenAI-compatible API.
​

Core Functionality
It automatically routes requests to the correct provider based on the model identifier (e.g., aws:anthropic.claude-3-haiku), handles authentication via environment variables or config, and standardizes request/response formats across providers like OpenAI, Anthropic, AWS Bedrock, and others.


```python
client=ai.Client()
```

```python
message=[
    {
        "role":"system","content":"You are a helpful assistant",
        "role":"user","content":"What is 2+2"
    }
]
```

```python
model="aws:anthropic.claude-3-haiku-20240307-v1:0"
```

The temperature parameter in aisuite's chat.completions.create() controls the randomness and creativity of LLM responses, typically ranging from 0 to 2.

How It Works
Lower values (0-0.3) produce focused, deterministic outputs by favoring high-probability tokens, ideal for factual answers or code generation. Higher values (0.7-1.0+) increase diversity and creativity but risk less coherent results

```python
response=client.chat.completions.create(
    model=model,
    messages=message,
    temperature=0.7,
    max_tokens=100
)
```

```python
print("For the question, ",message[0].get("content"))
```

```python
print("The response is ",response.choices[0].message.content)
```
